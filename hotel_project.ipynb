# **hotel booking project**

```python
df=pd.read_csv("hotel_booking.csv.crdownload")
```

Arrival by month and total bookings arrivals

```python
monthly_bookings = df.groupby('arrival_date_month')['hotel'].count().reset_index()
monthly_bookings
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='arrival_date_month', y='hotel', data=monthly_bookings)
plt.title('Monthly Booking Trends')
plt.xlabel('Arrival Month')
plt.ylabel('Number of Bookings')
plt.xticks(rotation=45)
plt.show()
```

# average lead time for different booking statuses

```python
lead_time_by_status = df.groupby('is_canceled')['lead_time'].mean().reset_index()
print(lead_time_by_status)
```

```python
plt.pie(lead_time_by_status['lead_time'], labels=lead_time_by_status['is_canceled'], autopct='%1.1f%%')
plt.title('Average Lead Time by Booking Status')
plt.show()
```

# lead time

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['lead_time'], bins=30)
plt.title('Distribution of Lead Time')
plt.xlabel('Lead Time (days)')
plt.ylabel('Frequency')
plt.show()
```

# booking preferences based on customer type

```python

customer_type_bookings = df.groupby('customer_type')['hotel'].count().reset_index()
print(customer_type_bookings)
```

```python
plt.figure(figsize=(6, 6))
plt.pie(customer_type_bookings['hotel'], labels=customer_type_bookings['customer_type'], autopct='%1.1f%%')
plt.title('Distribution of Customer Types')
plt.show()
```

# room types

```python
room_type_bookings = df.groupby('reserved_room_type')['hotel'].count().reset_index()
print(room_type_bookings)
```

```python
plt.bar(room_type_bookings['reserved_room_type'], room_type_bookings['hotel'])
plt.title('Room Type Bookings')
plt.xlabel('Room Type')
plt.bar(room_type_bookings['reserved_room_type'], room_type_bookings['hotel'])
plt.title('Room Type Bookings')
plt.xlabel('Room Type')
plt.ylabel('Number of Bookings')
for p in plt.gca().patches:
    plt.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')
plt.show()
```

# customers prefrences

```python
customer_preferences = df.groupby(['customer_type', 'reserved_room_type'])['hotel'].count().reset_index()
print(customer_preferences)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='customer_type', y='hotel', hue='reserved_room_type', data=customer_preferences)
plt.title('Customer Preferences for Room Types')
plt.show()
```

# price by month

```python
pricing_trends = df.groupby('arrival_date_month')['adr'].mean().reset_index()
print(pricing_trends)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='arrival_date_month', y='adr', data=pricing_trends)
plt.title('Pricing Trends by Month')
for p in plt.gca().patches:
    plt.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')
plt.show()
```

# price by room

```python
pricing_by_room_type = df.groupby('reserved_room_type')['adr'].mean().reset_index()
print(pricing_by_room_type)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='reserved_room_type', y='adr', data=pricing_by_room_type)
plt.title('Pricing Trends by Room Type')
for p in plt.gca().patches:
    plt.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')
plt.show()
```
