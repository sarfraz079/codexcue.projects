# attacks by cities


```python
attacks_by_city = df['City'].value_counts().head(10)

attacks_by_city
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='City', order=df['City'].value_counts().index[:10])
plt.title('Top 10 Cities with Most Suicide Bombing Attacks')
plt.xlabel('City')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 5), textcoords='offset points')
plt.show()
```

### targets of atacks

```python
main_targets = df['Target Type'].value_counts().head(5)
main_targets
```

```python
    plt.bar(main_targets.index, main_targets.values)
    plt.xlabel('Target Type')
    plt.ylabel('Number of Attacks')
    for i, v in enumerate(main_targets.values):
        plt.text(i, v, str(v), ha='center', va='bottom')
    plt.title('Main Targets of Suicide Bombing Attacks')
    plt.show()
```

# max kills

```python
Killed_Max=df['Killed Max'].max()
print(Killed_Max)
```

# attack over time

```python
yearly_attacks=df.groupby('Year')['Killed Max'].sum()
yearly_attacks
```

```python
plt.bar(yearly_attacks.index, yearly_attacks.values)
plt.xlabel('Year')
plt.ylabel('Total Killed Max')
plt.title('Total Killed Max per Year')
plt.show()
```

highly defected locations

```python
highly_casultes_location = df.groupby('Location')['Killed Max'].max().sort_values(ascending=False).head(1)
highly_casultes_location
```

# max defected province

```python
province_with_most_attacks = df['Province'].value_counts().idxmax()
province_with_most_attacks
```

```python
latitude_max = df['Latitude'].max()
latitude_max
```

max longitude by location

```python
max_longitude_by_location = df.groupby('Location')['Longitude'].max().sort_values(ascending=False).head(1)
max_longitude_by_location
```

max attacks by islamic festival

```python
max_blast_by_islamic_days = df.groupby('Islamic Date')['Killed Max'].max().sort_values(ascending=False).head(1)
max_blast_by_islamic_days
```

```python
attacks_by_holidays = df['Holiday Type'].value_counts().sort_values(ascending=False).head(5)
```

```python
plt.bar(attacks_by_holidays.index, attacks_by_holidays.values)
plt.xlabel('Holiday Type')
plt.ylabel('Number of Attacks')
plt.title('Number of Attacks by Holiday Type')
plt.show()
```
